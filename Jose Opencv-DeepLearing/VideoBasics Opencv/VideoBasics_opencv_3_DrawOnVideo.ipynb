```python
import cv2



width= int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height= int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
#Top left corner
x= width//2
y= height//2 # we divide like this because we want to stay as int
#width and height of Rectangle
w = width//4
h = height//4

#Bottom right x+w, y+h

#Callback func rectangle
def draw_rect(event,x,y,flags,param):
    global point1,point2,topLeft_clicked,botRight_clicked
    
    if event == cv2.EVENT_LBUTTONDOWN: # sol tika basildi mi kontrolu
        #reset the rectangle(it checks if the rect there)
        if topLeft_clicked == True and botRight_clicked== True: # sekil cizildikten sonra seklin cizildiginin kontrolu ve degerlerin sifirlanmasi
            point1 = (0,0)
            point2=(0,0)
            topLeft_clicked = False
            botRight_clicked = False
        if topLeft_clicked == False:
            point1 = (x,y)
            topLeft_clicked = True
        elif botRight_clicked == False:
            point2 = (x,y)
            botRight_clicked = True
            
            
            
# global variables
point1 = (0,0)
point2=(0,0)
topLeft_clicked = False
botRight_clicked = False
#connect to the callback

cap = cv2.VideoCapture(0)
cv2.namedWindow('Test')
cv2.setMouseCallback('Test',draw_rect)
while True:
    
    ret,frame = cap.read()
    
    #cv2.rectangle(frame,(x,y),(x+w,y+h),color = (0,0,255),thickness=4) -->burada videonun herhangi bir yerine bir dikdortgen cizdik
    if topLeft_clicked == True:
        cv2.circle(frame,center = point1,radius = 5,color = (0,0,255),thickness=-1)
    if topLeft_clicked == True and botRight_clicked == True:
        cv2.rectangle(frame,point1,point2,(0,0,255),thickness = 3)
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        

cap.release()
cv2.destroyAllWindows()
```
